```python
import pandas as pd
```

# 1. 0. Alibaba

```python
prediction_2600_alibaba = pd.read_csv("Export/alibaba/alibaba_prediction_variantB_2600_clusterwise.csv").drop(columns = ['Unnamed: 0'])
prediction_5000_alibaba = pd.read_csv("Export/alibaba/alibaba_prediction_variantB_5000_clusterwise.csv").drop(columns = ['Unnamed: 0'])
prediction_10000_alibaba = pd.read_csv("Export/alibaba/alibaba_prediction_variantB_10000_clusterwise.csv").drop(columns = ['Unnamed: 0'])
```

```python
all_predictions_alibaba = pd.DataFrame()
```

## 1.1 Preprocessing

```python
prediction_2600_alibaba['users_after_classifier']=2600
prediction_2600_alibaba['cluster_prediction']= prediction_2600_alibaba.cluster_id.apply(lambda x:x[:x.rindex('_')])
prediction_5000_alibaba['users_after_classifier']=5000
prediction_5000_alibaba['cluster_prediction']= prediction_5000_alibaba.cluster_id.apply(lambda x:x[:x.rindex('_')])
prediction_10000_alibaba['users_after_classifier']=10000
prediction_10000_alibaba['cluster_prediction']= prediction_10000_alibaba.cluster_id.apply(lambda x:x[:x.rindex('_')])
prediction_10000_alibaba.sample(3)
```

```python
all_predictions_alibaba = all_predictions_alibaba.append(prediction_2600_alibaba)
all_predictions_alibaba = all_predictions_alibaba.append(prediction_5000_alibaba)
all_predictions_alibaba = all_predictions_alibaba.append(prediction_10000_alibaba)
```

```python
all_predictions_alibaba.pivot_table(index=['model','step','method'], columns=['users_after_classifier','cluster_prediction'], values='RMSE').round(3)
```

## 1.2 Adding predictions without clusterwise

```python
users = 2600
all_predictions_alibaba_wo_clusterwise_2600 =   pd.read_csv("Export/alibaba/alibaba_prediction_both_methods_"+str(users)+"_.csv").drop(columns = ['Unnamed: 0'])
all_predictions_alibaba_wo_clusterwise_2600['users_after_classifier']=users
all_predictions_alibaba_wo_clusterwise_2600['cluster_prediction']= all_predictions_alibaba_wo_clusterwise_2600.cluster_id.apply(lambda x:x if (x=='none') else ('cluster_'+ x[-1:]))
```

```python
all_predictions_alibaba.pivot_table(index=['model','step','method'], columns=['users_after_classifier','cluster_prediction'], values='RMSE').round(3)
```

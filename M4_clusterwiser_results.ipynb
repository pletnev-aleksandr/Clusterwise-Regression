```python
import pandas as pd
import os
```

# M4

```python
all_files = [x  for x in os.listdir('Export/ICDM19') if 'final2' in x]
all_files
```

```python
li = []

for filename in all_files:
    df = pd.read_csv('Export/ICDM19/'+filename, index_col=None, header=0).drop(columns = ['Unnamed: 0'])
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame['cluster']=frame.apply(lambda x:x.cluster_id[:-2],axis =1 )
```

```python
frame.pivot_table(values= ['Score','reassignments_iterations'], index= ['frequency','model','horizon','cluster'],columns=['final_loss'])
```

## Alibaba

```python
users = 10000
```

```python
prediction_B =  "Export/alibaba/alibaba_prediction_variantB_"+str(users)+"_.csv"
prediction_A_theta = "Export/alibaba/alibaba_prediction_theta_variantA_"+str(users)+"_.csv"
prediction_A_trmf = "Export/alibaba/alibaba_prediction_trmf_variantA_"+str(users)+"_.csv"
# Path to save prediction 
prediction_both_methods = "Export/alibaba/alibaba_prediction_both_methods_"+str(users)+"_.csv"
```

```python
li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0).drop(columns = ['Unnamed: 0'])
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
```

```python
frame.to_csv(prediction_both_methods)
```
